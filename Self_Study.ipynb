```python
import matplotlib.pyplot as plt

# 初始化
w_no_momentum = 5.0     # 初始权重（普通GD）
w_momentum = 5.0        # 初始权重（带动量）
learning_rate = 0.1
decay_rate = 0.9        # 动量衰减率

velocity = 0            # 初始化动量（即 delta）
w_history_gd = []
w_history_momentum = []

# 梯度下降过程
for i in range(30):
    # 当前梯度
    grad = 2 * w_no_momentum

    # 标准梯度下降
    w_no_momentum -= learning_rate * grad
    w_history_gd.append(w_no_momentum)

    # 带动量的梯度下降
    grad_m = 2 * w_momentum
    velocity = decay_rate * velocity - learning_rate * grad_m
    w_momentum += velocity
    w_history_momentum.append(w_momentum)

# 可视化
plt.plot(w_history_gd, label='Gradient Descent')
plt.plot(w_history_momentum, label='Momentum')
plt.xlabel('Iteration')
plt.ylabel('Weight w')
plt.title('Optimization with and without Momentum')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 更深的目标函数和梯度
def loss(w):
    return 10 * w**4 - 30 * w**3 + 5

def grad(w):
    return 40 * w**3 - 90 * w**2

# 初始化
w_gd = 1.0
w_momentum = 1.0
learning_rate = 0.01
decay_rate = 0.9
velocity = 0

gd_path = []
momentum_path = []

# 迭代优化
for i in range(100):
    # 普通GD
    g1 = grad(w_gd)
    w_gd -= learning_rate * g1
    gd_path.append(w_gd)

    # 带动量
    g2 = grad(w_momentum)
    velocity = decay_rate * velocity - learning_rate * g2
    w_momentum += velocity
    momentum_path.append(w_momentum)

# 可视化
w_vals = np.linspace(-1, 4, 300)
loss_vals = loss(w_vals)

plt.plot(w_vals, loss_vals, label='Loss: $10w^4 - 30w^3 + 5$', color='gray')
plt.plot(gd_path, loss(np.array(gd_path)), label='Gradient Descent', marker='o', markersize=3)
plt.plot(momentum_path, loss(np.array(momentum_path)), label='Momentum', marker='x', markersize=3)
plt.xlabel('w')
plt.ylabel('Loss')
plt.title('Deeper Local Minima with and without Momentum')
plt.legend()
plt.grid(True)
plt.show()
```
